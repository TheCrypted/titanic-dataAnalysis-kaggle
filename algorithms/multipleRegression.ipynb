```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
X = np.array([1, 2, 3])
Y = np.array([20000, 30000, 40000])
```

```python
def predict(x, w, b):
    return x*w + b
```

```python
def findGradientStep(x, yhat, y):
    square = ((yhat - y))*x
    return np.mean(square)
```

```python
import random


def findBestWeights(x, y, alpha, num_iters):
    w = 0
    b = 0
    while True:
        estimateCurrent = predict(x, w, b)
        wTemp = w - alpha*(findGradientStep(x, estimateCurrent, y))
        bTemp = b - alpha*(findGradientStep(1, estimateCurrent, y))
        if abs(wTemp - w) == 0:
            return wTemp, bTemp
        w = wTemp
        b = bTemp
```

```python
findBestWeights(X, Y, 0.01, 100)
```
